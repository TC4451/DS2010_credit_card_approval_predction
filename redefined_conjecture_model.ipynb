```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("clean_dataset.csv")
y = df['Approved']
x = df['Employed']
graph = plt.plot(x,y)
sns.countplot(x='Employed', hue='Approved', data=df)
plt.xlabel("Employed (0=No, 1=Yes)")
plt.ylabel("Credit Card Approval Count")
plt.legend()
plt.show()
```
